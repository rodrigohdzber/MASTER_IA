# Sharpe Rolling optimizado

```python
import pandas as pd
import numpy as np
from time import time
from timeit import timeit
window = 10
```

```python
tiempo_inicial = time()

# Cargamos los datos y extraemos por un lado los activos, la renta fija y el índice Dax
datos=pd.read_csv('DAX.csv',sep=';')
activos=datos.drop(['Fecha', 'Dax', 'BUND'], 1) # Extraemos los activos quitando las últimas columnasy la primera.
BUND=datos.loc[:, 'BUND'] # Extraemos la renta fija.

# Generamos las matrices de datos donde guardaremos los resultados.
rent_activos=pd.DataFrame(np.zeros((activos.shape[0] , activos.shape[1])), index=activos.index, columns=activos.columns)
rent_bund=pd.DataFrame(np.zeros(len(BUND)))
sharpe_activos=pd.DataFrame(np.zeros((activos.shape[0] , activos.shape[1])), index=activos.index, columns=activos.columns)

# Calculamos la rentabilidad de los activos, renta fija e índice.
rent_activos = np.log(activos).diff()
rent_BUND = np.log(BUND).diff()
rent_activos.iloc[:1]=0
rent_BUND[0]=0
volatilidad=pd.DataFrame(np.zeros((activos.shape[0] , activos.shape[1])), index=activos.index, columns=activos.columns)

for activo in activos.columns:
    volatilidad.loc[:,activo] =(rent_activos.loc[:,activo].rolling(window).std())
    
volatilidad=volatilidad.fillna(0)
Rfactivos =pd.DataFrame(np.ones((rent_activos.shape[0] , rent_activos.shape[1])), index=rent_activos.index, columns=rent_activos.columns)
rent_BUND = Rfactivos.mul(rent_BUND, axis=0)
sharpe_activos = (rent_activos - rent_BUND) / volatilidad
sharpe_activos.replace([np.inf, -np.inf], np.nan, inplace = True)
sharpe_activos = sharpe_activos.fillna(0)
tiempo_final = time() 
tiempo_ejecucion = tiempo_final - tiempo_inicial

print ('El tiempo de ejecucion fue:',tiempo_ejecucion )

sharpe_activos.head(20)
```

## Otra manera de hacerlo

```python
import pandas as pd
import numpy as np
import time
```

```python
start_time = time.time()

filename = r'DAX.csv'
window = 10

# Read data and extract assets,bund(c[-1]),dax(c[-2])
pd_data = pd.read_csv(filename, header=0,
                      parse_dates=True, index_col=0, sep=';', dayfirst=True)

pd_data_log_ret_windowed = np.log(pd_data).diff()

rolling_var = np.log(pd_data).diff().rolling(window).std()

sharpe = pd_data_log_ret_windowed.iloc[:, :-1].add(
    pd_data_log_ret_windowed.iloc[:, -1].mul(
        -1, axis='index'), axis='index').div(
    rolling_var.iloc[:, :-1], axis='index')

print("--- %s seconds ---" % (time.time() - start_time))
```

## Otra manera de hacerlo

```python
import pandas as pd
import numpy as np
import time
from tqdm import tqdm
```

```python
start_time = time.time()

datos = pd.read_csv('DAX.csv', sep=';')
datos.set_index('Fecha', inplace=True)

ventana = 10

columns = datos.columns.tolist()
activos = datos[columns[:-2]]
BUND = datos[['BUND']]

rent_activos = np.log(activos).diff()
rent_BUND = np.log(BUND).diff()
volatilidad = rent_activos.rolling(ventana).var() ** 0.5
sharpe_activos = np.subtract(rent_activos, rent_BUND).iloc[ventana:] / volatilidad.iloc[ventana:]
sharpe_activos.replace([np.inf, -np.inf], 0, inplace=True)

print("--- %s seconds ---" % (time.time() - start_time))
```
