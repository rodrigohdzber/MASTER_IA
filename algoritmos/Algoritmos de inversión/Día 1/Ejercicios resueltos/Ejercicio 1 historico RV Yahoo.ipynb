```python
def get_data(ticker,start_date, end_date):
    
    stock_data = data.DataReader(ticker,'yahoo',start_date, end_date)    
    return stock_data   
```

```python
activos_a_descargar=("EMBR3.SA", "ENBR3.SA", "EQTL3.SA", "CPI.L", "CRH.L", "DC.L", "DCC.L", "DGE.L", "HEN3.DE", "IFX.DE","LHA.DE", "LIN.DE", "MRK.DE")
```

```python
def dowload_yahoo(activos,profundidad):
    """funcion que permite descargar datos
    de Yahoo desde una tupla, y pintar 
    los resultados en una lista de DataFrames.
    ----
    activos: tupla con tics
    profundidad: dias a recuperar"""
    
    lista_precios=[]
    for activo in activos:
        x=yf.Ticker(activo)
        df=x.history(period=profundidad)
        lista_precios.append(df)
        globals()[activo]=df.copy() #nos lo llevamos a un dataframe con el nombre del activo
        
    return lista_precios
```

```python
lista_precios=dowload_yahoo(activos_a_descargar,"5d")
```

La gracia de separarlo en dos funciones, es que con la segunda puedo recuperar la info mas adelante sin malgastar descargas

```python
def df_activo(serie_activos,lista_DF,activo):
    x=serie_activos.index(activo)
    df=lista_DF[x]
    return df  
```

```python
# Recupero EQTL3.SA
df_activo(activos_a_descargar,lista_precios,"EQTL3.SA")
```
